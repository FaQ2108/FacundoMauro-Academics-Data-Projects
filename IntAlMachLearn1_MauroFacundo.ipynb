b) En la columna Pclass, completar los valores NA, asignando a las personas a la clase 2.

```python
data1 = data.dropna()

data1 = 2

data['pclass'] = data['pclass'].fillna(data1)

muestra_aleatoria = data["pclass"].sample(n=5)
print(muestra_aleatoria)
```

```python
data.isna().sum()
```

c) En la clase â€œSexâ€, imputar los datos de manera de poder completar la base solo con los valores â€œmaleâ€ y â€œfemaleâ€.

```python
data['sex'] = data['sex'].apply(lambda x: np.random.choice(['male', 'female']) if pd.isnull(x) else x)

data2 = data[data['sex'].isin(['male', 'female'])]
muestra_aleatoria2 = data2["sex"].sample(n=5)

print(muestra_aleatoria2)
```

```python
data.isna().sum()
```

d) En los casos de las personas de las cuales no se tenga la edad, rellenar los datos con el promedio de edad de aquellos que sÃ­ tienen el dato.

```python
data['age'] = data['age'].fillna(data['age'].mean())

muestra_aleatoria3 = data["age"].sample(n=5)

print(muestra_aleatoria3)
```

```python
data.isna().sum()
```

e) En la columna embarked, asignar el valor â€œSâ€ a aquellos datos que se encuentran con valor vacÃ­o.

```python
data['embarked'] = data['embarked'].fillna('S')

muestra_aleatoria4 = data["embarked"].sample(n=5)

print(muestra_aleatoria4)
```
